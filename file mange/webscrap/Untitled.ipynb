```python
import requests
import bs4

URL = "https://quotes.toscrape.com/page/{}/"
HEADER = {'User-Agent': 'Mozilla/5.0'}  # Fixed typo in Mozilla
NUM = 1  # Start with page 1

while True:
    # Get the page
    base = requests.get(URL.format(str(NUM)), headers=HEADER, timeout=15)
    soup = bs4.BeautifulSoup(base.text, 'lxml')
    
    # Extract quotes and authors
    quotes = soup.select('.text')
    authors = soup.select('.author')
    
    # Check if we got any quotes
    if not quotes:
        print(f"No quotes found on page {NUM}. Ending.")
        break
    
    # Print quotes and authors
    print(f"--- Page {NUM} ---")
    for i in range(min(len(quotes), len(authors))):
        print(authors[i].text)
        print(quotes[i].text)
        print()
    
    # Check if there's a next page
    next_button = soup.select('.next')
    if not next_button:
        print("No more pages. Ending.")
        break
    
    # Move to next page
    NUM += 1
```
