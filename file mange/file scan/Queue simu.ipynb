```python
from collections import deque
import random

def customer_simulation(simulation_time=10, probability_of_arrival=0.35, seed=None):
    if seed is not None:
        random.seed(seed)
    
    customers = deque()
    next_id = 1
    wait_times = []
    max_customer = 0
    serving = False  # Flag to track if we're currently serving a customer
    
    print("Starting...... (Simulation bank queue)")
    for current_time in range(simulation_time):
        # Check for new arrivals
        if random.random() < probability_of_arrival:
            customers.append((next_id, current_time))
            print(f"[Time={current_time:02d}] arrives -> customer{next_id} -> (Queue={len(customers)})")
            next_id += 1  # Increment customer ID for next arrival
        else:
            print(f'[Time={current_time:02d}] no one arrives (Queue={len(customers)})')
        
        # Serve customer if not currently serving and queue is not empty
        if not serving and customers:
            serving = True  # Start serving
            customer_id, arrive_time = customers.popleft()
            wait = current_time - arrive_time
            wait_times.append(wait)
            print(f"[Time={current_time:02d}] serving customer {customer_id}, wait={wait}, queue={len(customers)}")
        else:
            # If we were serving, finish service after one time unit
            if serving:
                serving = False  # Finished serving
                print(f"[Time={current_time:02d}] finished serving customer, queue={len(customers)}")
            
        max_customer = max(max_customer, len(customers))
    
    # Print final statistics
    served = len(wait_times)
    avg_wait = sum(wait_times) / served if served else 0.0
    print("\nFinal Statistics:")
    print(f"Ticks: {simulation_time}")
    print(f'Served: {served}')
    print(f"Avg wait: {avg_wait:.2f}")
    print(f'Max queue: {max_customer}')

# Call the function correctly
customer_simulation()
```

```python
from collections import deque
import random
import time


def customer_simulation(simulation_time=10, probabily_of_arival=0.35, seed=None):
    """
    Simulates a bank queue with customers arriving randomly.
    
    Args:
        simulation_time: Total time units to run the simulation
        probabily_of_arival: Chance of a new customer arriving each time unit
        seed: Random seed for reproducibility
    
    Note: In this simulation, a customer is served immediately in the same time unit
    they arrive if no one else is in the queue. This doesn't account for service time.
    """
    # Set random seed if provided for reproducible results
    if seed is not None:
        random.seed(seed)
    
    # Initialize queue and tracking variables
    customer = deque()  # Queue to store customers (id, arrival_time)
    next_id = 1         # ID to assign to next customer
    wait_times = []     # List to track how long each customer waited
    max_customer = 0    # Track maximum queue length during simulation

    # Start simulation
    print("Stating...... (Simulation bank queue)")
    for num in range(simulation_time):
        # Check if a new customer arrives based on probability
        if random.random() < probabily_of_arival:
            # Add new customer to queue with (id, arrival_time)
            customer.append((next_id, num))
            
            print(f"[num={num:02d}] arives -> customer{next_id}-> (Queue={len(customer)})")
            next_id += 1  # Bug fix: increment customer ID
            
        else:
            # No customer arrives this time unit
            print(f'[Time={num:02d} no one arives (queue={len(customer)})')
        
        # Serve the next customer if queue is not empty
        # Bug: This serves a customer in the same time unit they arrive
        # Should consider adding service time to make simulation more realistic
        if customer:
            # Remove customer from front of queue
            customer_id, arive_time = customer.popleft()
            wait = num - arive_time  # Calculate wait time
            wait_times.append(wait)  # Record wait time
            print(f"severd  customer {customer_id} wait={wait} quw={len(customer)}")
            max_customer = max(max_customer, len(customer))  # Update max queue length

        # Print simulation statistics
        served = len(wait_times)
        avg_wait = sum(wait_times)/served if served else 0.0
        print(f"ticks: {simulation_time}")
        print(f'served: {served}')
        print(f"avg wait: {avg_wait:.2f}")
        print(f'max que :{max_customer}')
        
```

```python
from collections import deque
import random

def customer_simulation(simulation_time=50, arrival_p=0.4, seed=42, service_time_low=1, service_time_high=3):
    if seed is not None:
        random.seed(seed)

    queue = deque()              # (customer_id, arrival_tick)
    next_id = 1
    wait_times = []

    # server state
    serving = False
    remaining_service = 0
    current_customer = None
    busy_ticks = 0

    # stats
    max_queue = 0
    queue_len_sum = 0
    served = 0

    print("Starting... (Bank queue simulation)")

    for t in range(simulation_time):
        # --- arrivals ---
        if random.random() < arrival_p:
            queue.append((next_id, t))
            print(f"[t={t:02d}] arrival -> customer {next_id}  (queue={len(queue)})")
            next_id += 1
        else:
            print(f"[t={t:02d}] no arrival       (queue={len(queue)})")

        # --- if server is idle, start next service immediately ---
        if not serving and queue:
            current_customer, arrive_t = queue.popleft()
            wait = t - arrive_t
            wait_times.append(wait)
            remaining_service = random.randint(service_time_low, service_time_high)
            serving = True
            print(f"[t={t:02d}] START service -> customer {current_customer} (wait={wait}, svc={remaining_service} ticks)")

        # --- service progress for this tick ---
        if serving:
            busy_ticks += 1
            remaining_service -= 1
            if remaining_service == 0:
                # finish this tick
                print(f"[t={t:02d}] FINISH service -> customer {current_customer}")
                served += 1
                serving = False
                current_customer = None
                # IMPORTANT: try to start another customer in the SAME tick
                if queue:
                    current_customer, arrive_t = queue.popleft()
                    wait = t - arrive_t
                    wait_times.append(wait)
                    remaining_service = random.randint(service_time_low, service_time_high)
                    serving = True
                    print(f"[t={t:02d}] START service -> customer {current_customer} (wait={wait}, svc={remaining_service} ticks)")

        # --- stats per tick ---
        qlen = len(queue)
        queue_len_sum += qlen
        max_queue = max(max_queue, qlen)

    avg_wait = (sum(wait_times) / served) if served else 0.0
    avg_queue = queue_len_sum / simulation_time
    utilization = busy_ticks / simulation_time

    print("\nSummary")
    print(f"ticks:         {simulation_time}")
    print(f"served:        {served}")
    print(f"avg wait:      {avg_wait:.2f} ticks")
    print(f"avg queue len: {avg_queue:.2f}")
    print(f"max queue:     {max_queue}")
    print(f"utilization:   {utilization:.2f}")

# try it
customer_simulation(simulation_time=50, arrival_p=0.4, seed=7, service_time_low=1, service_time_high=3)
```
